```python
import glob
import numpy as np
import cv2
import os

from PIL import Image
from matplotlib import pyplot as plt
from tqdm.auto import tqdm
```

```python
path_whole_imgs = "./data/funny_birds/v2/test_interventions/**/**/body_beak_eye_foot_tail_wing.png"
path_beak = "./data/funny_birds/v2/test_interventions/**/**/body_eye_foot_tail_wing.png"
path_eye = "./data/funny_birds/v2/test_interventions/**/**/body_beak_foot_tail_wing.png"
path_foot = "./data/funny_birds/v2/test_interventions/**/**/body_beak_eye_tail_wing.png"
path_tail = "./data/funny_birds/v2/test_interventions/**/**/body_beak_eye_foot_wing.png"
path_wings = "./data/funny_birds/v2/test_interventions/**/**/body_beak_eye_foot_tail.png"
```

```python
paths_img = sorted(glob.glob(path_whole_imgs))

paths_b = sorted(glob.glob(path_beak))
paths_e = sorted(glob.glob(path_eye))
paths_f = sorted(glob.glob(path_foot))
paths_t = sorted(glob.glob(path_tail))
paths_w = sorted(glob.glob(path_wings))
```

```python
for p_img, p_b, p_e, p_f, p_t, p_w in zip(paths_img, paths_b, paths_e, paths_f, paths_t, paths_w):
    img_wi = np.array(Image.open(p_img))[:, :, :-1].astype(float)
    
    img_b = np.array(Image.open(p_b))[:, :, :-1].astype(float)
    img_e = np.array(Image.open(p_e))[:, :, :-1].astype(float)
    img_f = np.array(Image.open(p_f))[:, :, :-1].astype(float)
    img_t = np.array(Image.open(p_t))[:, :, :-1].astype(float)
    img_w = np.array(Image.open(p_w))[:, :, :-1].astype(float)
    
    diff_b = np.sum(abs(img_wi - img_b), axis=-1)
    diff_b = np.array(diff_b > 0).astype(np.uint8) * 1
    
    diff_e = np.sum(abs(img_wi - img_e), axis=-1)
    diff_e = np.array(diff_e > 0).astype(np.uint8) * 2
  
    diff_f = np.sum(abs(img_wi - img_f), axis=-1)
    diff_f = np.array(diff_f > 0).astype(np.uint8) * 4
    
    diff_t = np.sum(abs(img_wi - img_t), axis=-1)
    diff_t = np.array(diff_t > 0).astype(np.uint8) * 8
    
    diff_w = np.sum(abs(img_wi - img_w), axis=-1)
    diff_w = np.array(diff_w > 0).astype(np.uint8) * 16

    parts = diff_w + diff_t + diff_f + diff_e + diff_b

    p_class, p_img = p_img.split(os.path.sep)[-3:-1]

    os.makedirs(os.path.join(".", "new_parts", p_class), exist_ok=True)    
    cv2.imwrite(os.path.join(".", "new_parts", p_class, f"{p_img}.png"), parts,)
```

```python
# from matplotlib import pyplot as plt
plt.imshow(parts);
plt.axis('off');
```
